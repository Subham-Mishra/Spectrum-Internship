```python
from sklearn.preprocessing import LabelEncoder
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
```

```python
# Assigning numerical values and storing in another column
df['Pstatus_cat'] = labelencoder.fit_transform(df['Pstatus'])
df['school_cat'] = labelencoder.fit_transform(df['school'])
df['sex_cat'] = labelencoder.fit_transform(df['sex'])
df['address_cat'] = labelencoder.fit_transform(df['address'])
df['famsize_cat'] = labelencoder.fit_transform(df['famsize'])
df['Mjob_cat'] = labelencoder.fit_transform(df['Mjob'])
df['Fjob_cat'] = labelencoder.fit_transform(df['Fjob'])
df['reason_cat'] = labelencoder.fit_transform(df['reason'])
df['guardian_cat'] = labelencoder.fit_transform(df['guardian'])
df['famsup_cat'] = labelencoder.fit_transform(df['famsup'])
df['paid_cat'] = labelencoder.fit_transform(df['paid'])
df['activities_cat'] = labelencoder.fit_transform(df['activities'])
df['nursery_cat'] = labelencoder.fit_transform(df['nursery'])
df['higher_cat'] = labelencoder.fit_transform(df['higher'])
df['internet_cat'] = labelencoder.fit_transform(df['internet'])
df['romantic_cat'] = labelencoder.fit_transform(df['romantic'])
df
```

```python
df['final_grade']=df['G1']+df['G2']+df['G3']
df
```

```python
df2 = df.select_dtypes(exclude=['object'])
df2
```

```python
df2.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
df3=df2.drop('G3',axis=1)
df3.columns
X=df3.iloc[:, :-1].values
X
```

```python
y=df.iloc[:,-1].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn import linear_model
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
predictions[:10]
```

```python
print('Score:', model.score(X_test, y_test))
```

```python
## The line / model
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```
